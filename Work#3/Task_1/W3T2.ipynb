2. Загрузите данные который вы получили на предыдущем уроке путем скрейпинга сайта с помощью Buautiful Soup в MongoDB и создайте базу данных и коллекции для их хранения.

```python
from pymongo import MongoClient
import json
```

```python
# Для небольшой базы данных без разделения на более мелкие фрагменты
# Поключеие к DB
client = MongoClient("mongodb://localhost:27017/")

# Создание DB и создание колекции
db = client["W3T2_db"]
collect = db["storage_books"]

# Чтение файла
with open("data_site_books.json", 'r', encoding='utf-8') as file:
    try:
        data = json.load(file)
    except Exception as e:
            print(f"Error load data: {e}")

    # Вставляем данные в колекцию
collect.insert_many(data)

print('Data_load')
```
